```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
#load tha data 
data= pd.read_csv('environmental factors.csv')
data.head()
```

```python
#normalize the features using standarScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
#display
print(pd.DataFrame(data_scaled, columns=data.columns).head())
```

```python
#use the elbow method to find the optimal number of cluster
inertia=[]
k_range = range(1,11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

#plot the inertia values to find the "elbow"
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of cluster (k)')
plt.ylabel('Inertia')
plt.show()
```

```python
#applying k-mean clustering with k =7 
k=7
kmeans=KMeans(n_clusters=k, random_state=42)
data['cluster'] = kmeans.fit_predict(data_scaled)
#display few rows with cluster
print(data.head())
```

```python
#calculate silhouette score
sil_score = silhouette_score(data_scaled, data['cluster'])
print(f'Silhouette Score: :{sil_score}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#assuming data already contains the cluster labels(from k means clustering model)
#we will use 2 features to plot: 'Carbon_emissions' and 'pollution_level'
plt.figure(figsize=(8,6))
#create a scatter plot with the cluster labels
sns.scatterplot(x='carbon_emissions', y = 'pollution_level', hue='cluster',data=data, palette='viridis',s=100,alpha=0.7,edgecolor='k')

#add title and labels
plt.title('K-Means Clustering of enviroment factors')
plt.xlabel('Carbon Emissions')
plt.ylabel('Pollution Level')
plt.legend(title='Cluster', bbox_to_anchor=(1.05,1), loc='upper right')
plt.show()
```
